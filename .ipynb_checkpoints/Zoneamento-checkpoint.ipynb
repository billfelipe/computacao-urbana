```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import folium
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("zoneamento.csv" , sep =";", decimal =",", encoding = "UTF-8")
df
```

```python
df = df.loc[(df['TIPO'].str.contains('COMERCIO' , na=False))]

X = np.array(df)
plt.scatter(X[:,0], X[:,1], alpha=0.2, s=50)

m = folium.Map(location=[df.LATITUDE.mean(), df.LONGITUDE.mean()], zoom_start=13, 
               tiles='OpenStreet Map')
for _, row in df.iterrows():
    folium.CircleMarker(
        location=[row.LATITUDE, row.LONGITUDE],
        radius=row.QUANTIDADE/100,
        color='blue',
        fill=True,
        fill_colour='blue',
        opacity = row.QUANTIDADE/100
    ).add_to(m)

m
```

```python
df = df.loc[(df['TIPO'].str.contains('INDUSTRIA' , na=False))]

X = np.array(df)
plt.scatter(X[:,0], X[:,1], alpha=0.2, s=50)

m = folium.Map(location=[df.LATITUDE.mean(), df.LONGITUDE.mean()], zoom_start=13, 
               tiles='OpenStreet Map')
for _, row in df.iterrows():
    folium.CircleMarker(
        location=[row.LATITUDE, row.LONGITUDE],
        radius=row.QUANTIDADE/10,
        color='#FF7700',
        fill=True,
        fill_color='#FF7700',
        opacity = row.QUANTIDADE/10,
        fill_opacity = row.QUANTIDADE/10
    ).add_to(m)

m
```

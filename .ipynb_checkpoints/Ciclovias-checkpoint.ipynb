```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import folium
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("ciclovias.csv" , sep =";", decimal =",", encoding = "UTF-8")
df
```

```python
X = np.array(df)
plt.scatter(X[:,0], X[:,1], alpha=0.2, s=50)

m = folium.Map(location=[df.LATITUDE.mean(), df.LONGITUDE.mean()], zoom_start=13, 
               tiles='OpenStreet Map')
for _, row in df.iterrows():
    folium.CircleMarker(
        location=[row.LATITUDE, row.LONGITUDE],
        radius=1,
        color='#1787FE',
        fill=True,
        fill_colour='#1787FE'
    ).add_to(m)

m
```

```python
dados = pd.read_csv("ciclovias.csv" , sep =";", decimal =",", encoding = "UTF-8")
dados = dados[['LATITUDE' , 'LONGITUDE']]

dados = np.array(dados)

def dbscan(X, eps, min_samples):
    ss = StandardScaler()
    X = ss.fit_transform(X)
    db = DBSCAN(eps=eps, min_samples=min_samples)
    db.fit(X)
    y_pred = db.fit_predict(X)
    plt.scatter(X[:,0], X[:,1],c=y_pred, cmap='Paired')
    plt.title("DBSCAN")
    return db

db = dbscan(dados , .1 , 9)
labels = db.labels_

no_clusters = len(set(labels)) - (1 if -1 in labels else 0)
print(no_clusters)

labels = np.array(labels)
labels = pd.DataFrame(labels)

dados = pd.read_csv("ciclovias.csv" , sep =";", decimal =",", encoding = "UTF-8")
dados = dados[['LATITUDE' , 'LONGITUDE']]
dados['LABELS'] = labels

display(dados)

df = dados

df = df.groupby(
     ['LABELS']
 ).agg(
     LATITUDE = ('LATITUDE','mean'),
     LONGITUDE = ('LONGITUDE','mean'),
     QUANTIDADE = ('LABELS','count')
 ).reset_index()

df


#df = pd.DataFrame(data = db.labels_)
#df.to_csv("labels.csv", index= False)
```
